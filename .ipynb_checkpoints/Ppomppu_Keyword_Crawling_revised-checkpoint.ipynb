```python
from urllib.request import urlopen
from selenium import webdriver
from bs4 import BeautifulSoup
import re
```

```python
driver = webdriver.Chrome('C:/Users/kogas/Downloads/chromedriver')
driver.implicitly_wait(3)
url = 'http://www.ppomppu.co.kr/'
driver.get(url)

keyword = input('무엇을 검색하시겠습니까?')

search_input_element = driver.find_element_by_name("keyword")
search_input_element.send_keys(keyword)
search_button_element = driver.find_element_by_tag_name("button")
search_button_element.click()

pageSource = driver.page_source
bsObj = BeautifulSoup(pageSource, 'html.parser')
#for tag in bsObj.select('.content > .title > a'):
#    print(tag.text, tag['href'])
#    
page=[]
for tag in bsObj.select('.pagination > .page > a'):
    page.append(tag.text)
pnum = page[-2]
```

```python
count=0
for i in range(1,3):
    ppompage = 'http://www.ppomppu.co.kr/search_bbs.php?search_type=sub_memo&page_no='+str(i)+'&keyword=%EB%AA%B0%ED%85%8C%EC%9D%BC&page_size=20&bbs_id=&order_type=date&bbs_cate=1'
    html = urlopen(ppompage)
    bsObj_p = BeautifulSoup(html, 'html.parser')
    day_n_view = bsObj_p.find_all('p',{"class":"desc"})
    sss= []
    for d in day_n_view:
        kkk = d.get_text().strip().split(sep='\n')
        sss.append(kkk)
    for tag in bsObj_p.select('.content > .title > a'):
        print("글 제목 : ", tag.text, "\n링크 :", tag['href'])
        print("작성 게시판 : ",sss[count][0]," ",sss[count][1]," 작성 날짜 : ",sss[count][2])
        k = tag['href']
        num = re.search('no=(.+)&k',k)
        cnum = num.group(1)
        print("글 번호 :", cnum)
        html_c = urlopen(k)
        soup = BeautifulSoup(html_c, 'html.parser')
        title = soup.find_all("font", {"class" : "view_title2"})
        aisle = soup.find_all("font", {"class" : "view_cate"})
        name = soup.find_all('font', {"class" : "view_name"})
        #atrbute = soup.find_all('td', {"class" : "han"})
        contents = soup.find_all(class_='pic_bg')
        
        #for t in title:
        #    print("글 제목 : ", t.get_text())
        #    print("-"*3)
        for n in name:
            print("글 작성자 : ", n.get_text())
            print("-"*3)
        #for b in atrbute:
        #    print(b.get_text())
        #    print("-"*3)
        for a in aisle:
            print("글 분류 : ", a.get_text())
            print("글 내용 : ")
        for c in contents:
            print( c.get_text().strip())
        count=count+1
        print("===================================================")
    count=0
    print("========================================"+str(i)+"페이지========================================")
```
